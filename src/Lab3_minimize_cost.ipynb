```python
import tensorflow as tf
import matplotlib.pyplot as plt # 시각화 library
```

```python
X = [1, 2, 3]
Y = [1, 2, 3]

W_val = []
cost_val = []

# Cost function
@tf.function(input_signature=[
    tf.TensorSpec(shape=None, dtype=tf.float32),
])
def compute_cost(w):
  hypothesis = X * w
  return tf.reduce_mean(tf.square(hypothesis - Y))

for i in range(-30, 50):
  w = i * 0.1
  cost = compute_cost(w)
  W_val.append(w)
  cost_val.append(cost.numpy().item())

# Plot
plt.plot(W_val, cost_val)
plt.xlabel("W")
plt.ylabel("Cost")
plt.show()
```

```python
x_data = [1, 2, 3]
y_data = [1, 2, 3]

W = tf.Variable(tf.random.normal([1]), name='weight')
X = tf.Variable(0.0)
Y = tf.Variable(0.0)

W_val = []
cost_val = []

learning_rate = 0.1

# Cost function
def compute_cost():
  hypothesis = W * x_data
  return tf.reduce_mean(tf.square(hypothesis - y_data))

# Training loop
for step in range(21):
  with tf.GradientTape() as tape: # loss(cost)를 계산하는 과정을 기록
    # 함수를 사용한 cost function
    cost = compute_cost()

    # Cost function 직접 작성
    # hypothesis = W * x_data
    # cost = tf.reduce_mean(tf.square(hypothesis - y_data))

  gradient = tape.gradient(cost, [W])   # TF2: 자동 미분
  W.assign_sub(learning_rate * gradient[0])

  print(step, cost.numpy(), W.numpy())
  W_val.append(W.numpy().item())
  cost_val.append(cost.numpy().item())

# Plot
plt.scatter(W_val, cost_val)

for i, (x, y) in enumerate(zip(W_val, cost_val)):
    plt.text(x, y, str(i), fontsize=9, ha='right')

plt.xlabel("W")
plt.ylabel("Cost")
plt.show()
```

```python
X = [1, 2, 3]
Y = [1, 2, 3]

W_val = []
cost_val = []

W = tf.Variable(-3.0)

learning_rate = 0.1

optimizer = tf.keras.optimizers.SGD(learning_rate)

for step in range(100):
  with tf.GradientTape() as tape:
    hypothesis = W * x_data
    cost = tf.reduce_mean(tf.square(hypothesis - y_data))

  gradient = tape.gradient(cost, [W])
  optimizer.apply_gradients(zip(gradient, [W]))

  print(step, W.numpy(), cost.numpy())
  W_val.append(W.numpy().item())
  cost_val.append(cost.numpy().item())


# Plot
plt.scatter(W_val, cost_val)

for i, (x, y) in enumerate(zip(W_val, cost_val)):
    plt.text(x, y, str(i), fontsize=9, ha='right')

plt.xlabel("W")
plt.ylabel("Cost")
plt.show()
```

# Manual Gradient와 Automatic Gradient를 비교해본다.

- Manual gradient와 automatic gradient는 같은 logic으로 동작하므로, step에 따라 gradient의 변화가 동일해야 함.
- 즉, Automatic gradient가 의도한대로 미분을 통해 제대로 동작하는 것을 확인

```python
X = [1, 2, 3]
Y = [1, 2, 3]

W = tf.Variable(5.0)

optimizer = tf.keras.optimizers.SGD(0.01)

manual_hist = []
auto_hist = []
W_hist = []
cost_hist = []

for step in range(100):
  with tf.GradientTape() as tape:
    hypothesis = X * W
    cost = tf.reduce_mean(tf.square(hypothesis - Y))

  # Automatic gradient (equivalent to compute_gradients)
  auto_grad = tape.gradient(cost, [W])

  # Print manual-like gradient for comparison
  manual_grad = tf.reduce_mean((W * X - Y) * X) * 2

  # Apply gradients (equivalent to apply_gradients)
  optimizer.apply_gradients(zip(auto_grad, [W]))

  print(step, manual_grad.numpy(), W.numpy(), auto_grad[0].numpy(), cost.numpy())

  manual_hist.append(manual_grad.numpy().item())
  auto_hist.append(auto_grad[0].numpy().item())
  W_hist.append(W.numpy().item())
  cost_hist.append(cost.numpy().item())

# Plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(manual_hist, label="Manual Gradient")
plt.plot(auto_hist, '--', label="Auto Gradient")
plt.title("Gradient Comparison")
plt.xlabel("Step")
plt.ylabel("Gradient")
plt.legend()

plt.subplot(1,2,2)
plt.plot(W_hist, label="W")
plt.plot(cost_hist, label="Cost")
plt.title("Training Progress")
plt.xlabel("Step")
plt.legend()

plt.tight_layout()
plt.show()
```
